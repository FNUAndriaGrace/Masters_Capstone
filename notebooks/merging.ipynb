```python
import pandas as pd

def print_head_tail(df, n=5):
    """Prints the head and tail of a DataFrame."""
    print("Head:")
    display(df.head(n))  # Using display for better formatting in Jupyter
    print("\nTail:")
    display(df.tail(n))  # Using display for better formatting in Jupyter

# Load the datasets
df1 = pd.read_csv('/content/1_Bus metro_transport data.csv')
df2 = pd.read_csv('/content/2_public,retail spaces.csv')
df3 = pd.read_csv('/content/3_commercial,Retail.csv')
df4 = pd.read_csv('/content/4_Health, Residential.csv')

# Merge the dataframes on 'station_id'
merged_df = df1.merge(df2, on='station_id', how='outer') \
               .merge(df3, on='station_id', how='outer') \
               .merge(df4, on='station_id', how='outer') \


# Display the head and tail of the merged dataframe
print_head_tail(merged_df)
```

```python
import pandas as pd

# Load the datasets
df1 = pd.read_csv('/content/1_Bus metro_transport data.csv')  # Contains metro_500m, bus_500m
df2 = pd.read_csv('/content/2_public,retail spaces.csv')  # Contains Parks_500m, Libraries_500m, etc.
df3 = pd.read_csv('/content/3_commercial,Retail.csv')  # Contains Supermarkets, Malls, Restaurants, etc.
df4 = pd.read_csv('/content/4_Health, Residential.csv')  # Contains hospitals, residential, civic_buildings, etc.

# Force merge: we will use `pd.concat` to add the columns as separate ones
# Concatenate df2, df3, df4 side by side (ignore `station_id` here, just append columns)
df2 = df2.drop(columns=['station_id'])  # Drop 'station_id' from df2
df3 = df3.drop(columns=['station_id'])  # Drop 'station_id' from df3
df4 = df4.drop(columns=['station_id'])  # Drop 'station_id' from df4

# Concatenate the dataframes
merged_df = pd.concat([df1, df2, df3, df4], axis=1)

# Reset index for clean final result
merged_df = merged_df.reset_index(drop=True)

# Display the head and tail of the merged dataframe
print(merged_df.head())
print(merged_df.tail())
```
